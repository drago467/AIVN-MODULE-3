```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#%matplotlib qt

data = pd.read_csv('iris.csv')
print(data.shape)
```

```python
data['species'].unique()
```

```python
data["species"].value_counts()
```

```python
# Scatter Plots

fig = data[data.Species=='Setosa'].plot(kind='scatter',x='sepal.length',y='sepal.width',color='blue', label='Setosa', figsize= (10,6))
data[data.Species=='Versicolor'].plot(kind='scatter',x='sepal.length',y='sepal.width',color='orange', label='Versicolor',ax=fig)
data[data.Species=='Virginica'].plot(kind='scatter',x='sepal.length',y='sepal.width',color='green', label='Virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length vs Width")
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width",
              data=data, hue="species")
plt.show()
```

```python
sns.jointplot(x="petal_length", y="petal_width",
              data=data, hue="species")
plt.show()
```

```python
sns.stripplot(y='sepal_length', x='species',
              data=data, hue="species")
plt.show()
```

```python
sns.stripplot(y='petal_length', x='species',
              data=data, hue="species")
plt.show()
```

```python
sns.boxplot(x="species", y="petal_length",
            data=data, hue="species")
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_length",
            data=data, hue="species")
plt.show()
```

```python
data.hist(edgecolor='black', bins=25, figsize=(12,6))
plt.show()
```

```python
# shows the bivariate relation between each pair of features

sns.pairplot(data=data, hue="species")
plt.show()
```

```python
sns.heatmap(data.iloc[:,0:4].corr(),
            annot=True, fmt="f")
plt.show()
```
